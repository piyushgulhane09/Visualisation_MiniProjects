```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataframe = pd.read_csv(r"country_wise_latest.csv")
```

```python
dataframe.head()
```

FETCHING PARTICULAR STATES STARTING ALPHABETICAL 

```python
A_dataframe = dataframe[dataframe['Country/Region'].str.startswith('A')]
A_dataframe
```

RELATION BETWEEN CONFIRMED CASES & DEATHS 

```python
x = A_dataframe['Confirmed']
y = A_dataframe['Deaths']
```

```python
sns.relplot(x=x,y=y,data=A_dataframe,hue=A_dataframe['Country/Region'])
```

RELATION BETWEEN CONFIRMED CASES & RECOVERIES

```python
x = A_dataframe['Confirmed']
y = A_dataframe['Recovered']
```

```python
sns.relplot(x=x,y=y,data=A_dataframe,hue=A_dataframe['Country/Region'])
```

RELATION BETWEEN CONFIRMED CASES & ACTIVE 

```python
x = A_dataframe['Confirmed']
y = A_dataframe['Active']
```

```python
sns.relplot(x=x,y=y,data=A_dataframe,hue=A_dataframe['Country/Region'])
```
